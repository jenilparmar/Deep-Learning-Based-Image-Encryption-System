```python
import numpy as np
from PIL import Image
```

```python

img = Image.open(r"images.jpg")
img_array = np.array(img)
print(img_array.shape)  # (height, width, channels)
flat_arr = img_array.flatten()
```

```python
img_regian = flat_arr.reshape(img_array.shape)


img_regian = img_regian.astype('uint8')
```

```python


# Convert array to image
img_out = Image.fromarray(img_regian)

# Save as file
img_out.save("restored_image.png")
```
